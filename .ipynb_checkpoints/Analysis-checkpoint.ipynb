```python
def check_existence(detail, summary):
    incorrect = list()
    for i in range(len(detail)):
        if all(item in summary.iloc[i].tolist() for item in detail.iloc[i].tolist()): pass
        else: incorrect.append(detail.iloc[i].tolist())
    if len(incorrect) == 0: print("All data is correct!")
    elif len(incorrect) > 0: print(incorrect)
```

```python
check_existence(detail=detail, summary=summary)
```
